```python
# Q1. Load and Tokenize Text
# Task: Load text from the Brown Corpus (select any category), convert it into a string, and tokenize it into individual words using NLTK.

import nltk
from nltk.corpus import brown
from nltk.tokenize import word_tokenize

nltk.download('brown')
nltk.download('punkt_tab')


humorWords = brown.words(categories='humor')
humorSentence = " ".join(humorWords)

tokens = word_tokenize(humorSentence)
print(tokens)
```

```python
# Q2. Visualize Raw Word Frequencies
# Task: From the tokenized words, calculate the frequency distribution and visualize it using a bar chart and a word cloud (top 20 words).

import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.probability import FreqDist

frequencyDistribution = FreqDist()

for word in word_tokenize(humorSentence):
  if word.isalpha():
    frequencyDistribution[word.lower()]+=1


top20Words = frequencyDistribution.most_common(20)
print(top20Words)


# bar
words, counts = zip(*top20Words)
plt.bar(words,counts, color='grey')
plt.title("Top 20 Words in Humor Category")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()


print()

# wordcloud
wordcloud=WordCloud(width=800, height=800, background_color='white').generate_from_frequencies(dict(top20Words))
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```

```python
# Q3. Apply Stemming (PorterStemmer)
# Task: Apply NLTK’s PorterStemmer to the tokenized words, then list each stemmed word along with its frequency.

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmedWords = []

for word in tokens:
  if word.isalpha():
    stemmedWords.append(stemmer.stem(word.lower()))

frequencyDistributionStems = FreqDist(stemmedWords)

for word, freq in frequencyDistributionStems.items():
    print(word, freq)
```

```python
# Q4. Apply Lemmatization (WordNetLemmatizer)
# Task: Apply NLTK’s WordNetLemmatizer to the tokenized words, then list each lemmatized word along with its frequency.
from nltk.stem import WordNetLemmatizer
import pandas as pd

nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

lemmatizedWords = []

for word in tokens:
  lemmatizedWords.append(lemmatizer.lemmatize(word.lower()))

frequencyDistributionLemma = FreqDist(lemmatizedWords)

for word, freq in frequencyDistributionLemma.items():
    print(word, freq)
```

```python
# Q5. Compare Stemming vs Lemmatization
# Task: Create a side-by-side table or bar chart showing how stemming and lemmatization have reduced words differently (compare counts for at least 15 common words).

from collections import Counter

lemmaCounts = Counter(lemmatizedWords)
stemCounts = Counter(stemmedWords)

combinedCounts = Counter(lemmatizedWords + stemmedWords)

top15Words = []

for word, count in combinedCounts.most_common(15):
  top15Words.append(word)


data = {
    'Word': top15Words,
    'Lemmatized Count': [lemmaCounts[word] for word in top15Words],
    'Stemmed Count': [stemCounts[word] for word in top15Words]
}

dataFrame = pd.DataFrame(data)

print(dataFrame)

x = range(len(dataFrame['Word']))

plt.bar([i - 0.2 for i in x], dataFrame['Lemmatized Count'], width=0.4, label='Lemmatized')
plt.bar([i + 0.2 for i in x], dataFrame['Stemmed Count'], width=0.4, label='Stemmed')

plt.xticks(x, dataFrame['Word'], rotation=45)
plt.ylabel("Frequency")
plt.title("Stemming vs Lemmatization (Top 15 Words)")
plt.legend()
plt.tight_layout()
plt.show()

```
